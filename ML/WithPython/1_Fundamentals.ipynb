### ðŸ§Š Tensor Shape Manipulation - reshaping, stacking, squeezing and unsqueezing

```python
# Reshaping - reshapes a tensor to a defined shape

# View - Return a view of an imput tensor of certain shape but keep the same memory as the original tensor

# stacking - combine multiple tensors on top of each other (vstack) or side by side

# Squeeze - removes all `1` dimension from a tensor
# Unsqueeze - add a `1` dimension to a tensor at specific dim
# Permute - return a view of the input with dimensions permuted (swapped) in a certain way

x.shape

x_reshaped = x.reshape(1, 9) #but 2,9 won't be compatible as 2*9 = 18 and  not 9
x_reshaped, x
```

```python
# Changing the view (doesn't manupulate original tensor unlike reshape)
z = x.view(3 , 3)
z, x
```

```python
# Changing z changes x (shares same memory as the original input)
z[:, 0] = 5
z, x
```

```python
# Stack - combine multiple tensors on top of each other (vstack) or side by side
x_stacked = torch.stack([x,x,x,x], dim=0)
x_stacked
```

```python
# Squeeze - removes all single dimension from a tensor
x_squeezed = x_reshaped.squeeze()
x_squeezed
```

```python
# Unsqueeze - add a single dimension to a tensor at specific dim
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
x_unsqueezed
```

```python
# Permute - return a view of the input with dimensions permuted (index swapped) in a certain way

x_original = torch.rand(size=(244, 244, 3)) # [height, width, color_channels]
x_permuted = x_original.permute(2, 0, 1) # [color_channels, height, width]
```

### ðŸ§Š Reproducebility (trying to take random out of random)

How Neural network learns -> start with random numbers -> perform tensor operations -> update numbers to manke them better representation of the data -> repeat the process

To reduce randomness in our experiments, we can set a random seed. A random seed is a number that initializes the random number generator, ensuring that the sequence of random numbers generated is the same each time the code is run.


```python
# Create random tensors but reproducible tensors
random_tensor_A = torch.rand(3, 4)
random_tensor_B = torch.rand(3, 4)

random_tensor_A == random_tensor_B #false


RANDOM_SEED=42
torch.manual_seed(RANDOM_SEED)
random_tensor_c = torch.rand(3, 4)
torch.manual_seed(RANDOM_SEED) # must pass on each time
random_tensor_d = torch.rand(3, 4)

random_tensor_c == random_tensor_d #true
```

```python
# Move tensor to GPU ðŸŸ¢
tensor_on_gpu = tensor.to(device)
tensor_on_gpu
```

```python
# Move tensor back to CPU ðŸ”´
## if tensor is on GPU, it cant be transform to numPy as numpy only works on CPU

# tensor_on_gpu.numpy() # throws error
tensor_back_on_cpu = tensor_on_gpu.cpu().numpy()
print(tensor_back_on_cpu, tensor_back_on_cpu.device)
```
