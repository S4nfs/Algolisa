### üßä Manipulating Tensors

```python
#Add

tensor = torch.tensor([1,2,3])
tensor + 10 # alternative, torch.add(tensor, 10)
```

```python
# Multiply
tensor * 10
```

```python
# Subtract
tensor - 10
```

```python
# Division
```

```python
# Multiplication (http://matrixmultiplication.xyz/)
#### 1. Element-wise multiplications

print(tensor * tensor)
```

```python
#### 2. Matrix multiplication (dot product)
# Must Satifies some rules of multiplication of tensors otherwise throws errors:
# - Inner Dimensions must match:
#   (3, 2) @ (3, 2) ‚ùå
#       |_____|
#   (2, 3) @ (3, 2) ‚úîÔ∏è
#   (3, 2) @ (2, 3) ‚úîÔ∏è

# - The resulting matrix has the shape of **outer dimension**:
#   (2, 3) @ (3, 2)   -->  (2, 2) ‚úîÔ∏è


# %%time 
torch.matmul(tensor, tensor) # aliases.. tensor @ tensor or simply torch.mm(t, t)
```

### üßä Common Troubleshooting Errors: 

```python
# Shape errors
tensor_A = torch.tensor([[1, 2], [3, 4], [5, 6]])
tensor_A.shape
```

```python
tensor_B = torch.tensor([[7, 10], [8, 11], [9, 12]])
tensor_B.shape
```

```python
# Use **transpose**  - switches the axes or dimensions of a given tensor
tensor_fixed_B = tensor_B.T # and now it satisfies the shape tensor_A * tensor_B = [3, 2] * [2, 3]

tensor_A @ tensor_fixed_B
```

### üßä Tensor Aggregations - min, max, mean, sum etc

```python
# Create a tensor
x = torch.arange(1, 10, 1)
x, x.shape
```

```python
# min
torch.min(x) , 'or', x.min() 
```

```python
# max
torch.max(x) , 'or', x.max() 
```

```python
# mean - must be of float32
torch.mean(x.type(torch.float32)) , 'or', x.type(torch.float32).mean() 
```

```python
# sum
torch.sum(x) , 'or', x.sum() 
```

```python
# Finding the positional min and max 
x.argmin()
```

```python
x.argmax()
```

### üßä Tensor Shape Manipulation - reshaping, stacking, squeezing and unsqueezing

```python
# Reshaping - reshapes a tensor to a defined shape

# View - Return a view of an imput tensor of certain shape but keep the same memory as the original tensor

# stacking - combine multiple tensors on top of each other (vstack) or side by side

# Squeeze - removes all `1` dimension from a tensor
# Unsqueeze - add a `1` dimension from a tensor
# Permute - return a view of the input with dimensions permuted (swapped) in a certain way

x.shape

x_reshaped = x.reshape(1, 9) #but 2,9 won't be compatible as 2*9 = 18 and  not 9
x_reshaped, x
```

```python
# Changing the view ()
z = x.view(3 , 3)
z, x
```
